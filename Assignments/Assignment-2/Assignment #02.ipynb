Scenario 2: Student Grades

You are managing student grades using a dictionary. You need to calculate the average grade.take 
atleast 5 Students grades & then calculate the average

```python
print("Grade Calculation")
 
No_of_subjects = int(input("\nEnter Number of Subjects: "))

student_1 = input("\nEnter Name of student: ")
obtain_marks_of_std1 = 0
for marks in range(No_of_subjects):
    obtain_marks_of_std1 += int(input("Enter Marks: "))
    
student_2 = input("\nEnter Name of student: ")
obtain_marks_of_std2 = 0
for marks in range(No_of_subjects):
    obtain_marks_of_std2 += int(input("Enter Marks: "))
    
student_3 = input("\nEnter Name of student: ")
obtain_marks_of_std3 = 0
for marks in range(No_of_subjects):
    obtain_marks_of_std3 += int(input("Enter Marks: "))
    
student_4 = input("\nEnter Name of student: ")
obtain_marks_of_std4 = 0
for marks in range(No_of_subjects):
    obtain_marks_of_std4 += int(input("Enter Marks: "))
    
student_5 = input("\nEnter Name of student: ")
obtain_marks_of_std5 = 0
for marks in range(No_of_subjects):
    obtain_marks_of_std5 += int(input("Enter Marks: "))

Total_Marks = No_of_subjects*100
    
Average_Grade = {student_1: obtain_marks_of_std1,
                 student_2: obtain_marks_of_std2,
                 student_3: obtain_marks_of_std3,
                 student_4: obtain_marks_of_std4,
                 student_5: obtain_marks_of_std5}

print("\nGRADES OF STUDENTS:")
for key, value in Average_Grade.items():
    print("Percentage of", key ,"is", value*100 / Total_Marks)
    
For_Average_perc = 0
    
for values in Average_Grade.values():
    For_Average_perc += values*100 / Total_Marks 
    
Average_perc = For_Average_perc*100 / 500
print(f"\nAverage percentage is {Average_perc}. ")

if Average_perc > 80:
    print("\nAverage Grades of the Students is 'A+' ")
elif 70 < Average_perc < 80:
    print("\nAverage Grades of the Students is 'A' ")
elif 60 < Average_perc < 70:
    print("\nAverage Grades of the Students is 'B'")
elif 50 < Average_perc < 60:
    print("\nAverage Grades of the Students is 'C'")
elif 40 < Average_perc < 50:
    print("\nAverage Grades of the Students is 'D'")
else:
    print("\nAverage Grades of the Students is 'E'")
```

Scenario 3: Word Frequency Counter

You are given a list of words, and you need to count the frequency of each word.

```python
list_of_words = ["apple", "banana", "apple", "orange", "banana", "grape", "apple"]
words_count = {}
for word in list_of_words:
    if word in words_count:
        words_count[word] += 1
    else:
        words_count[word] = 1
        
for word, count in words_count.items():
    print(f"{word} : {count} times")
```

Scenario 3: Password Strength Checker

You are creating a program to check the strength of passwords based on certain criteria.
• Password should be at least 8 characters long.
• Password should contain at least one digit.
• Password should contain at least one letter.

```python
Password = input("Enter Password: ")
if len(Password) < 8:
    print("Password should be at least 8 character long.")
elif not any(char.isdigit() for char in Password):
    print("Password should contain at least one digit.")
elif not any(char.isalpha() for char in Password):
    print("Password should contain at least one letter.")
else:
    print("Password is strong.")
```

Scenario 04: Voting System (Annual Employee Recognition Awards)

You are developing a simple voting system for a contest. Users can vote for their Favorite option, and 
you need to count the votes.

```python
print("Voting System for Contest")

num_voters = int(input("\nEnter Number of voters: "))

candidates = ['Muzammil', 'Anas', 'Shehroz']
votes = {candidate: 0 for candidate in candidates}
                 
print("\nVoting for 'Employee of the Year'")

for _ in range(num_voters):
    print("\nCandidates:")
    print('Muzammil: @')
    print('Anas: #')
    print('Shehroz: $')
                 
    vote = input("Enter the Character corresponding to your vote: ")
                
    if vote == '@':
        selected_candidate1 = candidates[0]
        votes[selected_candidate1] += 1
        print('vote for Muzammil recorded.')
    elif vote == '#':
        selected_candidate2 = candidates[1]
        votes[selected_candidate2] += 1
        print('vote for Anas recorded.')
    elif vote == '$':
        selected_candidate3 = candidates[2]
        votes[selected_candidate3] += 1
        print('vote for Shehroz recorded.')
    else:
        print("Invalid vote. Please choose a valid candidate. ")
        
        
candidates_1 = ['Taha', 'Umer', 'Junaid']
votes1 = {candidate: 0 for candidate in candidates_1}

                 
print("\nVoting for 'Team player of the year'")

for _ in range(num_voters):
    print("\nCandidates:")
    print('Taha: !')
    print('Umer: %')
    print('Junaid: ~')
                 
    vote_1 = input("Enter the Character corresponding to your vote: ")
                
    if vote_1 == '!':
        selected_candidate_1 = candidates_1[0]
        votes1[selected_candidate_1] += 1
        print('vote for Taha recorded.')
    elif vote_1 == '%':
        selected_candidate_2 = candidates_1[1]
        votes1[selected_candidate_2] += 1
        print('vote for Umer recorded.')
    elif vote_1 == '~':
        selected_candidate_3 = candidates_1[2]
        votes1[selected_candidate_3] += 1
        print('vote for Junaid recorded.')
    else:
        print("Invalid vote. Please choose a valid candidate. ")

candidates_01 = ['Basid', 'Hamza', 'Ali']
votes_01 = {candidate: 0 for candidate in candidates_01}
        
print("\nVoting for 'Innovation Award'")

for _ in range(num_voters):
    print("\nCandidates:")
    print('Basid: ?')
    print('Hamza: <')
    print('Ali: >')
                 
    vote_02 = input("Enter the Character corresponding to your vote: ")
                
    if vote_02 == '?':
        selected_candidate_01 = candidates_01[0]
        votes_01[selected_candidate_01] += 1
        print('vote for Basid recorded.')
    elif vote_02 == '<':
        selected_candidate_02 = candidates_01[1]
        votes_01[selected_candidate_02] += 1
        print('vote for Hamza recorded.')
    elif vote_02 == '>':
        selected_candidate_03 = candidates_01[2]
        votes_01[selected_candidate_03] += 1
        print('vote for Ali recorded.')
    else:
        print("Invalid vote. Please choose a valid candidate. ")


print("\nVoting Results:")

print("\nEmployee of the year:")
for candidate, vote_count in votes.items():
    print(f"{candidate}: {vote_count} votes")

print("\nTeam player of the year:")
for candidate, vote_count in votes1.items():
    print(f"{candidate}: {vote_count} votes")
    
print("\nInnovation Award:")
for candidate, vote_count in votes_01.items():
    print(f"{candidate}: {vote_count} votes")
    
```
